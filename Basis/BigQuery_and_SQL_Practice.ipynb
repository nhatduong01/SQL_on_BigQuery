```python
from google.cloud import bigquery

# Create a "Client" object
client = bigquery.Client()
data_ref = client.dataset("nhtsa_traffic_fatalities", 
                          project="bigquery-public-data")
dataset = client.get_dataset(data_ref)

# Print all the tables in the dataset
tables = list(client.list_tables(dataset))

for table in tables:
  print(table.table_id)
```

```python
table_ref = data_ref.table('accident_2015')

table = client.get_table(table_ref)

# Print the table
client.list_rows(table, max_results = 5).to_dataframe()
```

### Determine the day in week which has the most accidents happened

```python
query = """
        SELECT COUNT(1) AS Num_accidents,
        day_of_week AS Day
        FROM `bigquery-public-data.nhtsa_traffic_fatalities.accident_2016`
        GROUP BY day_of_week
        ORDER BY Num_accidents DESC"""
query_job = client.query(query= query, project= 'bigquerysqlexercises')
accidents_by_days = query_job.to_dataframe()
accidents_by_days
```
